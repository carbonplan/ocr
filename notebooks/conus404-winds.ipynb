```python
fig, axes = plt.subplots(1, 2, figsize=(12, 9), constrained_layout=True)
catalog.get_dataset('conus404-ffwi-p95-wind-direction-mode').to_xarray().wind_direction_mode.plot(
    ax=axes[0], robust=True
)
catalog.get_dataset('conus404-ffwi-p99-wind-direction-mode').to_xarray().wind_direction_mode.plot(
    ax=axes[1], robust=True
)
plt.show()
```

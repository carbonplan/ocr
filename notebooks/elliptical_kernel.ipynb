```python
from ocr.risks.fire import generate_wind_directional_kernels

kernel_size = 41
circle_diameter = 35
weights_dict = generate_wind_directional_kernels(
    kernel_size=kernel_size, circle_diameter=circle_diameter
)
```

```python
import matplotlib.pyplot as plt

fig, axarr = plt.subplots()
axarr.imshow(weights_dict['E'], cmap='viridis')
axarr.set_title('Easterly wind')
axarr.scatter(20, 20, color='red')
plt.tight_layout()
```

```python
# confirm that the length of the non-zero axis is equal to the radius of the circle
# plus one (the +1 represents the central pixel in the filter)
(weights_dict['E'][kernel_size // 2, kernel_size // 2 :] > 0).sum() == circle_diameter // 2 + 1
```

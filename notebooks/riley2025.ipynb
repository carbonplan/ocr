```python
def load_fsim_results(path):
    print('starting')
    bp = rioxarray.open_rasterio(path+"/{}.tif".format('BP'))
    print('bp')
    ds = bp.to_dataset(name='BP')
    flp_vars = [f'FLP{length}' for length in np.arange(1,7)]
    for flp_var in flp_vars:
        print(flp_var)
        ds[flp_var] = rioxarray.open_rasterio(path+"/{}.tif".format(flp_var))
    ds = ds.chunk({'y': 6000, 'x': 4500})
    return ds
```

```python
historical_path = "s3://carbonplan-risks/riley2025/2011climaterun"
future_path = "s3://carbonplan-risks/riley2025/2047climaterun"
```

```python
recreate_riley_zarr = False
if recreate_riley_zarr:
    for path in [historical_path, future_path]:
        ds = load_fsim_results(path)
        ds.to_zarr(path+'.zarr', consolidated=True, mode='w')
else:
    riley = {}
    riley['2011'] = xr.open_zarr(historical_path+'.zarr')
    riley['2047'] = xr.open_zarr(future_path+'.zarr')
```

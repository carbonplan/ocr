```python
import xarray as xr
from distributed import Client

from ocr import catalog
from ocr.template import IcechunkConfig

client = Client()
client
```

```python
cr_2011 = catalog.get_dataset('2011-climate-run').to_xarray()
cr_2047 = catalog.get_dataset('2047-climate-run').to_xarray()

cr_2011_30m = catalog.get_dataset('2011-climate-run-30m-4326').to_xarray()
cr_2047_30m = catalog.get_dataset('2047-climate-run-30m-4326').to_xarray()

cr_30 = catalog.get_dataset('USFS-wildfire-risk-communities').to_xarray()


branch = 'QA'
icechunk_config = IcechunkConfig(branch=branch)
icechunk_repo_and_session = icechunk_config.repo_and_session(readonly=True)

QA = xr.open_zarr(icechunk_repo_and_session['session'].store, consolidated=False, chunks={})
print('2011')
print(float(cr_2011['BP'].min()))
print(float(cr_2011['BP'].max()))
print('\n')
print('2047')
print(float(cr_2047['BP'].min()))
print(float(cr_2047['BP'].max()))
print('\n')

print('2011-30m')
print(float(cr_2011_30m['BP'].min()))
print(float(cr_2011_30m['BP'].max()))
print('\n')

print('2047-30m')

print(float(cr_2047_30m['BP'].min()))
print(float(cr_2047_30m['BP'].max()))
print('\n')

print('community_risk_30m')
print(float(cr_30['BP'].min()))
print(float(cr_30['BP'].max()))
print(float(cr_30['CRPS'].min()))
print(float(cr_30['CRPS'].max()))
print('\n')
print('QA')
print('risk_2011')
print(float(QA['risk_2011'].min()))
print(float(QA['risk_2011'].max()))
print('\n')
print('risk_2047')
print(float(QA['risk_2047'].min()))
print(float(QA['risk_2047'].max()))
print('\n')
print('wind_risk_2011')
print(float(QA['wind_risk_2011'].min()))
print(float(QA['wind_risk_2011'].max()))
print('\n')
print('wind_risk_2047')
print(float(QA['wind_risk_2047'].min()))
print(float(QA['wind_risk_2047'].max()))
```
